```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/density 17A.csv')
```

```python
data.head()
```

```python
data.drop(index=[0,1,2,3,5])
```

```python
data.info()
```

```python
data.rename(columns={'DEPTH: 0':'Column1'},inplace=True)
```

```python

data.isnull().sum()
```

```python
data.isna().sum()
```

```python
columns_to_fill=[ 'Column3',	'Column4',	'Column5']
for i in columns_to_fill:
  if data[i].dtype in ['int64','float64'] :
     data[i]=data[i].fillna(data[i].mean())
else:
     data[i]=data[i].fillna(data[i].mode())
```

```python
ax1 = sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# Convert relevant columns to numeric, handling non-numeric values
for column in data.columns:
    if data[column].dtype == 'object':
        try:
            data[column] = pd.to_numeric(data[column], errors='coerce')  # Convert to numeric, replacing non-convertible values with NaN
        except:
            pass  # Handle cases where conversion is not possible

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# prompt: write a code to generate the well logged data plot

import matplotlib.pyplot as plt

# Create a figure and axes object
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.plot(data['Column1'], data['Column2'], color='black', linewidth=1)

# Add labels and title
ax.set_xlabel('Depth (m)', fontsize=14)
ax.set_ylabel('Density (g/cm3)', fontsize=14)
ax.set_title('Well Logged Data Plot', fontsize=16)

# Add grid lines
ax.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: write a code to generate the well logged data plot in each columns

# Loop through each column and create a separate plot
for column in data.columns:
    if column not in ['Column1', 'Column2']:  # Skip the first two columns
        # Create a figure and axes object
        fig, ax = plt.subplots(figsize=(10, 6))

        # Plot the data
        ax.plot(data['Column1'], data[column], color='black', linewidth=1)

        # Add labels and title
        ax.set_xlabel('Depth (m)', fontsize=14)
        ax.set_ylabel(column, fontsize=14)
        ax.set_title(f'Well Logged Data Plot for {column}', fontsize=16)

        # Add grid lines
        ax.grid(True)

        # Show the plot
        plt.show()
```

```python

import matplotlib.pyplot as plt
# Create a figure and subplots
fig, axs = plt.subplots(1, 4, figsize=(25, 30) , sharey=True)
df = data
# Plot each log on a separate subplot
axs[0].plot(df['Column2'], df['Column1'])
axs[0].set_xlabel('hrom')
axs[0].set_ylabel('Depth')
axs[0].invert_yaxis()  # Invert y-axis for well log convention

axs[1].plot(df['Column3'], df['Column1'])
axs[1].set_xlabel('hlef')

axs[2].plot(df['Column4'], df['Column1'])
axs[2].set_xlabel('hbdc')

axs[3].plot(df['Column5'], df['Column1'])
axs[3].set_xlabel('hdeb')

# Adjust spacing and display the plot
plt.tight_layout()
plt.show()
```
